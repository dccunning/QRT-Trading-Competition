## Quantatative Trading Strategy Notebook

### Breakdown of Portfolio Returns


$$
R_t = \alpha + \beta_1 F_{1,t} + \beta_2 F_{2,t} + \dots + \beta_k F_{k,t} + \epsilon_t
$$
Where:
- $R_t$ = strategy return
- $F_{i,t}$ = known risk factors (e.g. Market, HML, MOM, etc.)
- $\beta_i$ = exposure to those factors
- $\alpha$ = average return not explained by those factors
- $\epsilon_t$ = noise (unexplained time-varying residual)

### Trading Strategy Plan

1. Multi-factor base (momentum + Value + Quality)

2. Signal research — find additional alpha sources

3. Combine signals into composite score

4. Optimize portfolio on composite score

### Tradeable Factors

1. Momentum: buy winners, sell losers
    - Why: Information slow to spread; people hold onto losers and sell winners; short-sell constraints
    - How: t-12 to t-1 month return to predict month t returns
    - Data: Prices
2.	Value: buy cheap, sell expensive
    - Why: Investors overpay for growth and neglect distressed or boring firms; extrapolation bias; limits to arbitrage prevent prices from correcting instantly
    - How: Rank stocks on valuation ratios (e.g. low P/B, low P/E, low EV/EBITDA), long cheapest quantile, short most expensive
    - Data: Accounting fundamentals + prices
3.	Quality: buy strong businesses, sell weak ones
    - Why: Investors chase narratives and ignore balance-sheet strength; junk firms are overpriced lottery tickets; leverage risk underpriced in good times
    - How: Rank on profitability and balance-sheet metrics (e.g. high ROE, high gross profit/asset, low debt, stable earnings), long highest quality, short lowest quality
    - Data: Accounting fundamentals
